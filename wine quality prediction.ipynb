# Convert into Classification problem

```python
data['good_quality']=[1 if x>=7 else 0 for x in data["quality"]]
```

```python
data.head()
```

```python
data['good_quality'].value_counts()
```

```python
data['good_quality'].value_counts()
```

```python
X= data.drop(["quality","good_quality"],axis=1)
Y=data["good_quality"]
```

```python
from sklearn.preprocessing import StandardScaler
old_X_features= X
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.20, random_state=0)
```

```python
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=1)
dt.fit(X_train, y_train)
y_pred1 = dt.predict(X_test)
print(classification_report(y_test, y_pred1))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred2 = rf.predict(X_test)
print(classification_report(y_test, y_pred2))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred2)
```

```python
confusion_matrix = pd.crosstab(y_test, y_pred2, rownames=['Actual'], colnames=['Predicted'])
sb.heatmap(confusion_matrix, annot=True)
plt.show()
```

```python
import xgboost as xgb
xgb = xgb.XGBClassifier(random_state=1)
xgb.fit(X_train, y_train)
y_pred5 = xgb.predict(X_test)
print(classification_report(y_test, y_pred5))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(random_state=1)
gb.fit(X_train, y_train)
y_pred4 = gb.predict(X_test)
print(classification_report(y_test, y_pred4))
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(X_train,y_train)
y_pred6=lr.predict(X_test)
print(classification_report(y_test, y_pred6))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred3=knn.predict(X_test)
print(classification_report(y_test, y_pred3))
```

```python
feat_importances = pd.Series(rf.feature_importances_, index=old_X_features.columns)
feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```
